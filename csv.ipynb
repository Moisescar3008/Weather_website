```python
import os

output_folder = "weather_charts"
os.makedirs(output_folder, exist_ok=True)  # Crea la carpeta si no existe

# Función para guardar gráficos individuales
def save_plot(fig, filename):
    path = os.path.join(output_folder, filename)
    fig.savefig(path, dpi=300, bbox_inches="tight")
    plt.close(fig)
    print(f"Gráfico guardado en: {path}")

# Crear la carpeta principal para almacenar los gráficos
main_output_folder = "weather_charts"
os.makedirs(main_output_folder, exist_ok=True)

# Función para guardar gráficos individuales
def save_plot(fig, folder, filename):
    path = os.path.join(folder, filename)
    fig.savefig(path, dpi=300, bbox_inches="tight")
    plt.close(fig)
    print(f"Gráfico guardado en: {path}")

# Obtener la lista de archivos CSV
csv_folder = "csv"
csv_files = [f for f in os.listdir(csv_folder) if f.endswith("_weather.csv")]

for file in csv_files:
    city_name = file.replace("_weather.csv", "")
    city_folder = os.path.join(main_output_folder, city_name)
    os.makedirs(city_folder, exist_ok=True)
    
    file_path = os.path.join(csv_folder, file)
    df = pd.read_csv(file_path)
    df["DateTime"] = pd.to_datetime(df["DateTime"])

    # Gráfico de temperatura
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Temperature (C)"], marker="o", linestyle="-", color="r", label="Temperatura (°C)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Temperatura (°C)")
    ax.set_title("Temperatura en los próximos días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "temperature.png")

    # Gráfico de humedad
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Humidity"], marker="s", linestyle="-", color="b", label="Humedad (%)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Humedad (%)")
    ax.set_title("Humedad en los próximos días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "humidity.png")

    # Gráfico de velocidad del viento
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Wind speed"], marker="^", linestyle="-", color="g", label="Velocidad del Viento (m/s)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Velocidad del Viento (m/s)")
    ax.set_title("Velocidad del viento en los próximos días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "wind_speed.png")

    # Gráfico de presión atmosférica
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Pressure"], marker="o", linestyle="-", color="purple", label="Presión (hPa)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Presión (hPa)")
    ax.set_title("Presión Atmosférica en los Próximos Días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "pressure.png")

    # Gráfico de nubosidad
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Cloudiness"], marker="s", linestyle="-", color="gray", label="Nubosidad (%)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Nubosidad (%)")
    ax.set_title("Nubosidad en los Próximos Días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "cloudiness.png")

    # Gráfico de precipitación
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Rain volume"], marker="v", linestyle="-", color="blue", label="Lluvia (mm)")
    ax.plot(df["DateTime"], df["Snow volume"], marker="^", linestyle="-", color="cyan", label="Nieve (mm)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Precipitación (mm)")
    ax.set_title("Precipitación en los Próximos Días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "precipitation.png")

    # Gráfico de probabilidad de precipitación
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["DateTime"], df["Rain volume"] * 100, marker="v", linestyle="-", color="blue", label="Lluvia (%)")
    ax.plot(df["DateTime"], df["Snow volume"] * 100, marker="^", linestyle="-", color="cyan", label="Nieve (%)")
    ax.set_xlabel("Fecha y Hora")
    ax.set_ylabel("Probabilidad de Precipitación (%)")
    ax.set_title("Probabilidad de Precipitación en los Próximos Días")
    ax.legend()
    ax.grid()
    ax.tick_params(axis="x", rotation=45)
    save_plot(fig, city_folder, "precipitation_probability.png")

    # Gráfico de rosa de los vientos
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, polar=True)
    wind_directions = np.deg2rad(df["Wind direction"])
    wind_speeds = df["Wind speed"]
    ax.scatter(wind_directions, wind_speeds, c="g", alpha=0.75, label="Velocidad del viento (m/s)")
    ax.set_title("Rosa de los Vientos")
    ax.legend()
    save_plot(fig, city_folder, "wind_rose.png")

print("Todos los gráficos han sido guardados en la carpeta 'weather_charts' con subcarpetas por ciudad.")

```
